**The Problem**

The problem that you will be solving is called the SELECTION problem.

**SELECTION**

Input A sequence of n ≥ 2 numbers and an integer k, 1 ≤ k ≤ n

Output The k
th smallest number in the sequence.

**Approach Used**  
Based on the hint given I thought about the Quicksort Algorithim that was covered in class. This algorithim was also talked at length in the DataStructures and Algorithims course. Quicksort is a sorting algorithim that selects a pivot postion and sorts all elements less than that in left array and all elements greater in the right array. This is the same approach that I used to solve the Selection Problem. What is unique about the Selection Problem is that it wants the kth smallest number, not a sorted array or the smallest element.

```python
def selection(array, k):
    if len(array) == 1:
        return array[0]

    pivot = array[0]
    left = []
    right = []
    for i in range(1,len(array)):
        if array[i] < pivot:
            left.append(array[i])
        else:
            right.append(array[i])

    if k <= len(left):
        return selection(left, k)
    elif k == len(left) + 1:
        return pivot
    else:
        return selection(right, k - (len(left)+1))

print(selection([5, 1, 6, 7, 3, 4, 8], 3))
```

**Recurrence Relation Derivation**

**Assumptions:**
- We assume the pivot divides the elements approximately in half each time (best/average case)
- Let T(n) be the running time for an array of size n

**Analysis:**

At each recursive call, the algorithm performs the following operations:

1. **Partitioning Work:** We iterate through n-1 elements (excluding the pivot) to partition them into left and right arrays. This takes Θ(n) time.

2. **Recursive Call:** After partitioning, we make exactly ONE recursive call on either the left or right subarray (not both, unlike quicksort).

**Step-by-step derivation:**

Given that the pivot divides the array in half:
- Size of left subarray ≈ n/2
- Size of right subarray ≈ n/2

The recurrence relation is:

**T(n) = T(n/2) + cn**

Where:
- **T(n/2)** represents the recursive call on half the array
- **cn** represents the linear work to partition the array (c is a constant)
