**SELECTION**

Input A sequence of n ≥ 2 numbers and an integer k, 1 ≤ k ≤ n

Output The k
th smallest number in the sequence.

**Approach Used**  
Based on the hint given I thought about the Quicksort Algorithim that was covered in class. This algorithim was also talked at length in the DataStructures and Algorithims course. Quicksort is a sorting algorithim that selects a pivot postion and sorts all elements less than that in left array and all elements greater in the right array. This is the same approach that I used to solve the Selection Problem. What is unique about the Selection Problem is that it wants the kth smallest number, not a sorted array or the smallest element.

```python
def selection(array, k):
    if len(array) == 1:
        return array[0]

    pivot = array[0]
    left = []
    right = []
    for i in range(1,len(array)):
        if array[i] < pivot:
            left.append(array[i])
        else:
            right.append(array[i])

    if k <= len(left):
        return selection(left, k)
    elif k == len(left) + 1:
        return pivot
    else:
        return selection(right, k - (len(left)+1))

print(selection([5, 1, 6, 7, 3, 4, 8], 3))
```

**Explaination / Justification**  
The algorithim selection starts by selecting a pivot, the first element in array. If there is only one element in the array that will triggger the base return case where the only element in the array is returned.Everything smaller than the pivot will go into the left array and everything larger will go into the right array. Due to the way the pivot splits the array we know that left length + 1 equals k and the only element possible to be kth smallest is the pivot. Otherwise the algorithim is recursively called on left or right sides depending on if k greater left lengh. If k is greater then search right if not search left.  

Algorithim produces expected output stated in lab for example when the array [5,1,6,7,3,4,8] is inserted and k is chosen to be 3, the algorithim produces the number 4. This is the 3rd smallest in in the array. The same could be said for the base case. The base case (n = 1) is always true because the first element of array will be kth smallest as long as 1 is less  than or equal to k and k is less than or equal to n.  

The time complexity is at least O(n) as the work done to spilt left and right sides is O(n) since you must go through each element. After the inital work is done each half is only partially sorting on the pivot. For example if you sort all smaller than pivot and all greater, and you end up going on the left side then no more sorting work will ever be done on the right side.However O(n) is only the case if the pivot actually splits the array into around half. If that is not the case it make be less efficent.
